# eduAgent dataset

```python
import pandas as pd

file_path = './data/gkt_bert/dataset_gkt_bert_embed_test_public_past5.csv'

df1 = pd.read_csv(file_path, delimiter='\t')

print(df1.head())
```

```python
max(df1.groupby(['course_name', 'slide_id']).ngroup() + 1), max(df1.groupby(['course_name', 'question_id']).ngroup() + 1)
```

```python
max(pd.factorize(df1[['course_name', 'slide_id']].apply(tuple, axis=1))[0] + 1), max(pd.factorize(df1[['course_name', 'question_id']].apply(tuple, axis=1))[0] + 1)
```

```python
df1['unique_slide_id'] = df1.groupby(['course_name', 'slide_id']).ngroup() + 1
```

```python
df1['unique_question_id'] = df1.groupby(['course_name', 'question_id']).ngroup() + 1
```

```python
df1['correctness'] = (df1['student_choice'] == df1['correct_choice']).astype(int)
```

```python
df1.to_csv('./data/gkt_bert/eduAgent_emb_test.csv', index=False)
```
